```python
#importing data as data frame and importing libraries
import scipy.stats as stats
import numpy as np
import pandas as pd

df = pd.read_csv('wyniki.csv')
df
```

```python
#printing variance value
print(np.var(df))
```

```python
#replacing categorical data with numerical values
df.replace('M',0, inplace = True)
df.replace('F',1, inplace = True)
df
```

```python
a = df['plec']
b = df['ocena_koncowa']
```

```python
#performing a t test
stats.ttest_ind(a,b)
```
